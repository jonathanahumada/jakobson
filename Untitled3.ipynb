```python
from jakobson import results
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
import matplotlib.pyplot as pyplot
```

```python
c1 = results.leer_muestra('c1','./resultados')
```

```python
scaler = MinMaxScaler()
```

```python
c1['indice_metafora']= c1['metafora']/c1['w']
```

```python
c1[['metafora_n', 'metonimia_n']]= scaler.fit_transform(c1[['metafora','metonimia']])
```

```python
import seaborn as sns
```

```python
fig = pyplot.figure(figsize = (15,12))
plt = sns.barplot(x="categoria", y="metafora_n", data=c1,  palette="ch:r=-.2,d=.3_r")
pyplot.xticks(rotation=45)
```

```python
fig.savefig('c1_metafora.png')
```

```python
pyplot.show()
```
