```python
semantica = pd.read_csv("matriz_semantica.csv", index_col=0, names=["palabra","vector_semantico"])
```

```python
uso = pd.read_csv("matriz_uso.csv", index_col=0, names=["palabra","vector_uso"])
```

```python
valores = pd.read_csv("referencia_metaforica.csv", index_col=0, names=["palabra","taza"])
```

```python
valores
```

```python
result  =  pd.merge(semantica,  uso, on="palabra")
```

```python
consolidado = pd.merge(result, valores, on="palabra")
```

```python
consolidado.taza
```

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler()
```

```python
scaler.fit(consolidado.taza)
```

```python
type(consolidado.taza)
```

```python
consolidado
```

```python
scaler.fit(consolidado[["taza"]])
```

```python
scaler.mean_
```

```python
res_standard_scaler = pd.DataFrame(scaler.transform(consolidado[['taza']]), columns=["min_max"], index=consolidado.index)
```

```python
res_standard_scaler
```

```python
consolidado.min_max = res_standard_scaler
```

```python
consolidado["z"] = res_standard_scaler["min_max"]
```

```python
consolidado
```

```python
consolidado.plot()
```

```python
plt.show()
```

```python

min_max_scaler = preprocessing.MinMaxScaler()
```

```python
res_min_max_scaler = pd.DataFrame(min_max_scaler.fit_transform(consolidado[['taza']]), columns=["min_max"], index=consolidado.index)
```

```python
consolidado["min_max"] = res_min_max_scaler["min_max"]
```

```python
consolidado
```

```python
consolidado = consolidado.drop(columns=["<"])
```

```python
consolidado.plot()
```

```python
plt.show()
```
