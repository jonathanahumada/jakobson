```python
import nltk
import jakobson

import pprint
```

```python
mensaje = nltk.corpus.brown.words("cj22") #The Emerging Nations

r1 = nltk.corpus.brown.words("cj23") # J22	Max F. Millikan & Donald L. Blackmer, editors	The Emerging Nations
r2 = nltk.corpus.brown.words("cj24") # J24	Howard J. Parad	Preventive Casework: Problems & Implications
r3 = nltk.corpus.brown.words("cj25") # J25	Sister Claire M. Sawyer	Some Aspects of Fertility of a Tri-Racial Isolate
r4 = nltk.corpus.brown.words("cj26") # J26	Frank Lorimer	Demographic Information on Tropical Africa
r5 = nltk.corpus.brown.words("cj27") # J28	William H. Ittelson & Samuel B. Kutash, editors	Perceptual Changes in Psychopathology
r6 = nltk.corpus.brown.words("cj28") # J30	Raymond J. Corsini	Roleplaying in Business & Industry

insumos_referencia = [r1, r2, r3, r4, r5, r6]
corpora_referencia = []
```

```python
for mensaje in insumos_referencia:
    for w in mensaje:
        corpora_referencia.append(w)


    
```

Antes de obtener el *indice metafórico* se hacen calculos preliminares. 


```python
pprint.pprint(jakobson.matriz_semantica(mensaje))
```

Una *matriz semántica* está compuesta de *vectores semanticos*. El vector semántico es el concepto que representa el *eje de la seleccion* para una palabra w en el mensaje.

Una breve explicación del eje de la selección:

>Seleccion:
>La selección estudia qué palabra selecciona un hablante entre las palabras existentes de la lengua, más o menos similares y hasta cierto punto equivalentes. La selección se basa en la sinonimia o antonimia de una palabra. En otros términos, en su semántica

Por ejemplo, veamos cuál sería el vector semántico de "provincial":

```python
provincial = jakobson.vector_semantico("provincial")
provincial
```

```python
jakobson.vector_semantico("Hypothesis")
```

¿Cómo se calcula el vector semántico? A través de *Wordnet*. 

>WordNet® is a large lexical database of English. Nouns, verbs, adjectives and adverbs are grouped into sets of cognitive synonyms (synsets), each expressing a distinct concept. Synsets are interlinked by means of conceptual-semantic and lexical relations

los requerimientos del para el vector semantico eran que:
- representaran relaciones entre ideas
- fuera lo suficientemente extenso para buscar una gran cantidad de palabras, pertenecientes a distintos corpus.

Por eso se selecciono Wordnet.

Internamenta, Wordnet lemmatiza los palabra "provincial" y la compara con otros lemmas segun criterios linguísticos que más tarde podrían extender la funcionalidad de  `jakobson`. Lo más importante de un vector semántico, es que procura capturar una 'idea', más allá del mero lemma.
 

Ahora, luego de formar la matriz semántica para un mensaje, `jakobson` calcula la *matriz de uso*. Para cada *vector semántico*, se produce un *vector de uso*. El objetivo del vector de uso es representar --valga la redundancia-- el uso de cada componente de del vector semántico. ¿Cómo podríamos representar el uso de cada palabra? Asignandole un número. Se utilizó un modelo de Bag of words. Veamos un ejemplo:

```python
from nltk.probability import FreqDist
#bag of words para el corpora_referencia
fd = FreqDist(corpora_referencia)

provincial_u = jakobson.vector_uso(provincial, fd)
provincial_u
```

```python
print(fd["bucolic"])
print(fd["peasant"])
print(fd["provincial"])
```

Esto nos muestra que la idea denotada por "vector_semantico('provincial')" es bastante *escaza* en nuestro corpora de referencia.Notemos que el corpora de referencia es conjunto de  palabras mucho más grande que el de nuestro mensaje. 
Por lo que podemos decir que nuestro mensaje inicial está haciendo un uso del lenguaje más "metaforico".

Veamos, pues, como se vería la *matriz de uso*

```python

jakobson.matriz_uso(jakobson.matriz_semantica(mensaje), fd)
```

El *vector de uso* es la expresión cuantitativa del *vector semantico*. Procura capturar cualitativamente cuánto se usa (que tam frecuente) es la idea en un corpus. Comop ejemplo. Pod
