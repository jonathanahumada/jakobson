Una *matriz sem√°ntica* est√° compuesta de *vectores semanticos*. El vector sem√°ntico es el concepto que representa el *eje de la seleccion* para una palabra w en el mensaje.

Una breve explicaci√≥n del eje de la selecci√≥n:

>Seleccion:
>La selecci√≥n estudia qu√© palabra selecciona un hablante entre las palabras existentes de la lengua, m√°s o menos similares y hasta cierto punto equivalentes. La selecci√≥n se basa en la sinonimia o antonimia de una palabra. En otros t√©rminos, en su sem√°ntica

Por ejemplo, veamos cu√°l ser√≠a el vector sem√°ntico de "provincial":

```python
provincial = jakobson.vector_semantico("provincial")
provincial
```

```python
jakobson.vector_semantico("Hypothesis")
```

¬øC√≥mo se calcula el vector sem√°ntico? A trav√©s de *Wordnet*. 

>WordNet¬Æ is a large lexical database of English. Nouns, verbs, adjectives and adverbs are grouped into sets of cognitive synonyms (synsets), each expressing a distinct concept. Synsets are interlinked by means of conceptual-semantic and lexical relations

los requerimientos del para el vector semantico eran que:
- representaran relaciones entre ideas
- fuera lo suficientemente extenso para buscar una gran cantidad de palabras, pertenecientes a distintos corpus.

Por eso se selecciono Wordnet.

Internamenta, Wordnet lemmatiza los palabra "provincial" y la compara con otros lemmas segun criterios lingu√≠sticos que m√°s tarde podr√≠an extender la funcionalidad de  `jakobson`. Lo m√°s importante de un vector sem√°ntico, es que procura capturar una 'idea', m√°s all√° del mero lemma.
 

Ahora, luego de formar la matriz sem√°ntica para un mensaje, `jakobson` calcula la *matriz de uso*. Para cada *vector sem√°ntico*, se produce un *vector de uso*. El objetivo del vector de uso es representar --valga la redundancia-- el uso de cada componente de del vector sem√°ntico. ¬øC√≥mo podr√≠amos representar el uso de cada palabra? Asignandole un n√∫mero. Se utiliz√≥ un modelo de Bag of words. Veamos un ejemplo:

```python
from nltk.probability import FreqDist
#bag of words para el corpora_referencia
fd = FreqDist(corpora_referencia)

provincial_u = jakobson.vector_uso(provincial, fd)
provincial_u
```

```python
print(fd["bucolic"])
print(fd["peasant"])
print(fd["provincial"])
```

Esto nos muestra que la idea denotada por "vector_semantico('provincial')" es bastante *escaza* en nuestro corpora de referencia.Notemos que el corpora de referencia es conjunto de  palabras mucho m√°s grande que el de nuestro mensaje. 
Por lo que podemos decir que nuestro mensaje inicial est√° haciendo un uso del lenguaje m√°s "metaforico".

Veamos, pues, como se ver√≠a la *matriz de uso*

```python

jakobson.matriz_uso(jakobson.matriz_semantica(mensaje), fd)
```
