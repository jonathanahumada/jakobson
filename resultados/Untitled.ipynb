```python
def metafora_y_metonimia(data):
    """Hace un catplot """
    plot = sns.catplot(x='value', y ='categoria',data=data, kind='bar', col='variable')
    return plot
```

```python
def pivotear_indice(df):
    """llama melt() para poder hacer un gr√°fico faceteado. la variable suelta es metafora / metonimia y el valor es la cuantitativa"
    """
    melted = df.melt(id_vars=['categoria','metafora', 'metonimia', 'w','metacategoria'])
    return melted
    
```

```python
def pivotear_metacategorias(df):
    """Primero agrupa un df segun metacategoria y luego pivotea los indices"""
    por_metacategoria = df.groupby('metacategoria' , as_index=False).mean()
    return por_metacategoria.melt(id_vars=['metacategoria', 'metafora', 'metonimia','w'])
    
    
    
    
```
